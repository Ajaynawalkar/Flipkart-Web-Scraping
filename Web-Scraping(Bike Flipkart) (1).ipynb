```python
model=soup.findAll('div',{'class':'_4ddWXP'})
model
```

```python
vehicle_name= []
for i in model:
    p=i.text
    vehicle_name.append(p)
print(vehicle_name)
```

```python
vehicle_rating=soup.findAll('div',{'class':'_3LWZlK'})
vehicle_rating
```

```python
product_rating= []
for i in vehicle_rating:
    q=i.text
    product_rating.append(q)
print(product_rating)
```

```python
len(product_rating)
```

```python
vehicle_price=soup.findAll('div',{'class':'_30jeq3'})
vehicle_price
```

```python
product_price= []
for i in vehicle_price:
    q=i.text
    product_price.append(q)
print(product_price)
```

```python
len(product_price)
```

- We have Product Name, Product Rating, Product Price of the Vehicle section in the flipkart.

```python
product_rating=product_rating[0:40]
product_price=product_price[0:40]
```

```python
print(len(product_price))
print(len(product_rating))
print(len(vehicle_name))
```

## Visualisation

```python
plt.bar(Vehicle_data['Price'], Vehicle_data['Rating'])
plt.title('Price vs Rating', fontsize=16, color='r')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
